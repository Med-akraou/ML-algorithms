```python
#Sigmoid function
def sigmoid(z):
    """
    Compute the sigmoid of z

    Args:
        z (ndarray): A scalar, numpy array of any size.

    Returns:
        g (ndarray): sigmoid(z), with the same shape as z
         
    """ 
    g = 1/(1+np.exp(-z))
    return g
```

```python
def compute_cost(X, y, w, b):
    """
    Computes the cost over all examples
    Args:
      X : (ndarray Shape (m,n)) data, m examples by n features
      y : (array_like Shape (m,)) target value 
      w : (array_like Shape (n,)) Values of parameters of the model      
      b : scalar Values of bias parameter of the model
      lambda_: unused placeholder
    Returns:
      total_cost: (scalar)         cost 
    """
    
    pos =(y[:,0]==1)
    neg =(y[:,0]==0)
    loss1 = -np.sum(np.log(sigmoid(X[pos,:].dot(w)+b)))
    loss2 = -np.sum(np.log(1-sigmoid(X[neg,:].dot(w)+b)))
    loss = loss1+loss2
    #loss_sum = (np.sum(np.log(sigmoid(X[y>0]))) +np.sum(sigmoid(np.log(1-X(y==0))))
    return loss/len(y)
```

```python
def compute_gradient(X, y, w, b): 
    """
    Computes the gradient for logistic regression 
 
    Args:
      X : (ndarray Shape (m,n)) variable such as house size 
      y : (array_like Shape (m,1)) actual value 
      w : (array_like Shape (n,1)) values of parameters of the model      
      b : (scalar)                 value of parameter of the model 
      lambda_: unused placeholder.
    Returns
      dj_dw: (array_like Shape (n,1)) The gradient of the cost w.r.t. the parameters w. 
      dj_db: (scalar)                The gradient of the cost w.r.t. the parameter b. 
    """
    m=X.shape[0]
    dw = (1/m)*X.T.dot(sigmoid(X.dot(w)+b)-y)
    db = (1/m)*np.sum(sigmoid(X.dot(w)+b)-y)
    return dw,db
```

```python
#Gradient decente
import math
def gradient_descent(X, y, w_in, b_in,alpha, num_iters, lambda_): 
    """
    Performs batch gradient descent to learn theta. Updates theta by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      X :    (array_like Shape (m, n)
      y :    (array_like Shape (m,))
      w_in : (array_like Shape (n,))  Initial values of parameters
      b_in : (scalar)                 Initial value of parameter 
      alpha : (float)                 Learning rate
      num_iters : (int)               number of iterations to run gradient descent
      lambda_ (scalar, float)         regularization constant
      
    Returns:
      w : (array_like Shape (n,)) Updated values of parameters of the model after
          running gradient descent
      b : (scalar)                Updated value of parameter of the model after
          running gradient descent
    """
    
    # number of training examples
    m = len(X)
    
    # An array to store cost J and w's at each iteration primarily for graphing later
    J_history = []
    w_history = []
    
    for i in range(num_iters):
        
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0 or i == (num_iters-1):
            w_history.append(w_in[0][0])
            J_history.append(compute_cost(X, y, w_in, b_in))
            
        # Calculate the gradient and update the parameters
        dj_dw, dj_db = compute_gradient(X, y, w_in, b_in)   

        # Update Parameters using w, b, alpha and gradient
        w_in = w_in - alpha * dj_dw               
        b_in = b_in - alpha * dj_db              
        
    return w_in, b_in, J_history, w_history #return w and J,w history for graphing
```

```python
w_start = np.ones((2,1))
b_start=1
w_in, b_in, J_history, w_history = gradient_descent(X ,y, w_start, b_start, 0.5, 1000,None)
plt.plot(np.arange(len(J_history)),J_history)
plt.show()
```

```python
def predict(X):
    """
    X: vector of features or matrix for miltuple predections
    """
    Z = X.dot(w_in) + b_in
    A = 1 / (1 + np.exp(-Z))
    return A>=0.5
```
